```python
import requests
import getpass
import json

print("Step 1: Authenticate")
print("Enter your API key (the one from the NEW project):")
GEMINI_API_KEY = getpass.getpass("Key: ")

# 1. Get the list of models available to THIS specific key
print("\nStep 2: Fetching your available models...")
list_url = f"https://generativelanguage.googleapis.com/v1beta/models?key={GEMINI_API_KEY}"
list_response = requests.get(list_url)

if list_response.status_code != 200:
    print(f"‚ùå Critical Error: Could not verify key. Status: {list_response.status_code}")
    print(list_response.text)
else:
    my_models = list_response.json().get('models', [])
    print(f"‚úÖ Found {len(my_models)} models available to you.")

    # 2. Iterate and Test
    print("\nStep 3: Testing models for free quota...")
    found_working = False
    
    # We prioritize "Flash" models as they are most likely to be free
    # Sort specifically to put "flash" models first in line
    my_models.sort(key=lambda x: "flash" not in x['name'])

    for model_info in my_models:
        name = model_info['name'] # e.g., models/gemini-1.5-flash
        short_name = name.replace("models/", "")
        
        # Skip models that don't generate text (like embedding models)
        if "generateContent" not in model_info.get('supportedGenerationMethods', []):
            continue

        print(f"   Testing: {short_name}...", end=" ")
        
        # Construct the test URL
        test_url = f"https://generativelanguage.googleapis.com/v1beta/{name}:generateContent?key={GEMINI_API_KEY}"
        headers = {'Content-Type': 'application/json'}
        data = {"contents": [{"parts": [{"text": "Hello"}]}]}
        
        try:
            test_response = requests.post(test_url, headers=headers, json=data)
            
            if test_response.status_code == 200:
                print("‚úÖ SUCCESS!")
                print("\n" + "="*40)
                print(f"üéâ WE FOUND IT! YOUR WORKING MODEL IS: \"{short_name}\"")
                print("="*40)
                print(f"Response: {test_response.json()['candidates'][0]['content']['parts'][0]['text']}")
                found_working = True
                break # Stop the loop, we found the winner
            elif test_response.status_code == 429:
                print("‚ùå (Quota Exceeded)")
            else:
                print(f"‚ùå (Error {test_response.status_code})")
                
        except Exception as e:
            print(f"‚ùå (Network Error)")

    if not found_working:
        print("\n‚ùå Result: No working models found. This confirms the project has 0 quota globally.")
```
